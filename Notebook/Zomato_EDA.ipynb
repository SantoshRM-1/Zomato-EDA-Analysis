## Merging the two datasets colums

```python
final_df=pd.merge(df,df_country,on="Country Code",how='left')
final_df.head()
```

```python
df.head()
```

```python
#to check datatypes
final_df.dtypes
```

```python
country_names=final_df.Country.value_counts().index
country_names
```

```python
country_value=final_df.Country.value_counts().values
```

### Pie chart top 3 conuntries that use zomato

```python
plt.pie(country_value[:3],labels=country_names[:3],autopct='%1.2f%%')
```

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'rating count'})
```

```python
ratings.head()
```

## Obsevatiion
 1.when the ratiing id between 4.5 to 4.9-->excellent
 2. 4.0 to 4.4 --> very good
 3. 3.5 to 3.9--> average
 

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",
            y="rating count",
            data=ratings,
            hue="Rating color",
            palette=['brown','red','orange','yellow','green','blue'])
```

### with this graph we can make a observation
like 
1.not rated counts are very high
2.Maximun Number of ratings sre between 2.5 to 3.4

```python
## counts plot
sns.countplot(x="Rating color",
              data=ratings,
              hue="Rating color")
              # palette=['brown','red','orange','yellow','green','blue'])
```
